**t-test**

```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
df = pd.read_csv("/content/Covid Data.csv")
```

```python
df.head()
```

```python
print(df.columns)
```

```python
data = df["AGE"].dropna().values
print("Data:", data)
```

```python
# Define the null hypothesis
H0 = "The average AGE is 32."
H1 = "The average AGE is more than 32."

print("Null Hypothesis:", H0)
print("Alternative Hypothesis:", H1)
```

```python
# Calculate the test statistic comparing mean AGE to 32
t_stat, p_value_two_tailed = stats.ttest_1samp(data, 32)

print("T-Statistic:", t_stat)
print("P-Value (Two-Tailed):", p_value_two_tailed)
```

```python
# Convert two-tailed p-value to one-tailed (because H1 is "greater than 32")
p_value = p_value_two_tailed / 2

print("T-Statistic:", t_stat)
print("Two-Tailed P-Value:", p_value_two_tailed)
print("One-Tailed P-Value:", p_value)
```

```python
# Print results
print("Test statistic:", t_stat)
print("One-tailed p-value:", p_value)
```

```python
# Conclusion
if t_stat > 0 and p_value < 0.05:
    print("Reject the null hypothesis.")
else:
    print("Fail to reject the null hypothesis.")
```

**Two sampled T-test**

```python
from scipy.stats import ttest_ind
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Covid Data.csv")
```

```python
# Select groups based on PNEUMONIA status
# First, let's check the unique values in 'PNEUMONIA' to ensure the filtering condition is correct.
print(df['PNEUMONIA'].unique())

# Assuming 2 represents 'No Pneumonia' based on common encoding practices for binary categories (1 for Yes, 2 for No, 97/99 for missing/not applicable)
group1 = df[df["PNEUMONIA"] == 2]["AGE"].dropna().values   # No Pneumonia
group2 = df[df["PNEUMONIA"] == 1]["AGE"].dropna().values   # Pneumonia positive
print("Group 1 data (No Pneumonia):")
print(group1)
print("Group 2 data (Pneumonia Positive):")
print(group2)
```

```python
# Means
group1_mean = np.mean(group1)
group2_mean = np.mean(group2)

print("Group 1 mean value:", group1_mean)
print("Group 2 mean value:", group2_mean)
```

```python
# Standard deviations
group1_std = np.std(group1)
group2_std = np.std(group2)
print("group1 std value:", group1_std)
print("group2 std value:", group2_std)
```

```python
# Independent t-test
ttest, pval = ttest_ind(group1, group2, equal_var=False)  # safer for unequal sample sizes
print("p-value:", pval)
```

```python
# Conclusion
if pval < 0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```

**Paired sampled t-test**

```python
import pandas as pd
from scipy import stats
```

```python
# Describe AGE and PNEUMONIA
print(df[['AGE', 'PNEUMONIA']].describe())
```

```python
# Paired t-test (Before vs After for same customers)
ttest, pval = stats.ttest_rel(df['AGE'], df['PNEUMONIA'], nan_policy='omit')
print("p-value:", pval)
```

```python
# Conclusion
if pval < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

You would use a Z test if:

Your sample size is greater than 30. Otherwise, use a t test.

Data points should be independent from each other. In other words, one data point isn’t related or doesn’t affect another data point.

Your data should be normally distributed. However, for large sample sizes (over 30) this doesn’t always matter.

Your data should be randomly selected from a population, where each item has an equal chance of being selected.

Sample sizes should be equal if at all possible.


```python
import pandas as pd
from scipy import stats
from statsmodels.stats import weightstats as stests

# Load dataset
df = pd.read_csv("Covid Data.csv")

# Run one-sample Z-test on AGE
ztest, pval = stests.ztest(df['AGE'].dropna(),
                           x2=None,
                           value=32)  # hypothesized mean

print("p-value:", float(pval))
# Conclusion
if pval < 0.05:
    print("Reject null hypothesis")
else:
    print("Fail to reject null hypothesis")
```

**Two-sample Z test**

```python
import pandas as pd
from statsmodels.stats import weightstats as stests

# Load dataset
df = pd.read_csv("Covid Data.csv")

# Paired Z-test: Sales_Before vs Sales_After
ztest, pval1 = stests.ztest(df['AGE'].dropna(),
                            x2=df['PNEUMONIA'].dropna(),
                            value=0,
                            alternative='two-sided')

print(float(pval1))

# Conclusion
if pval1 < 0.05:
    print("reject null hypothesis")
else:
    print("accept null hypothesis")
```

**Chi-Square Test**

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load your dataset
df = pd.read_csv('Covid Data.csv')

# Create a contingency table
contingency_table = pd.crosstab(df['PATIENT_TYPE'], df['PNEUMONIA'])

# Perform the chi-square test
chi2_statistic, p_value, dof, expected_frequencies = chi2_contingency(contingency_table)

# Print the results
print('Chi-square statistic:', chi2_statistic)
print('P-value:', p_value)
print('Degrees of freedom:', dof)
print('Expected frequencies:')
print(expected_frequencies)
```
